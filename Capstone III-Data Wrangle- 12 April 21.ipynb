```python
import string
# Removing Punctuation
def remove_punctuation (text): 
    no_punct="".join([c for c in text if c not in string.punctuation])
    return no_punct
```

```python
df['reviewText']=df['reviewText'].apply(lambda x: remove_punctuation(x))
df['reviewText'].head()
```

```python
#Tokenization
from nltk.tokenize import RegexpTokenizer
tokenizer=RegexpTokenizer('\\s+')
df['reviewText']=df['reviewText'].apply(nltk.word_tokenize)
df['reviewText'].head()
```

```python
# removing StopWords
def remove_stopwords(text):
    words=[w for w in text if w not in stopwords.words('english')]
    return words
```

```python
df['reviewText']=df['reviewText'].apply(lambda x: remove_stopwords(x))
df['reviewText'].head()
```

```python
#Lemmatizer
from nltk.stem import WordNetLemmatizer

lemmatizer=WordNetLemmatizer()

def word_lemmatizer(text):
    lem_text=[lemmatizer.lemmatize(i) for i in text]
    return lem_text
```

```python
df['reviewText']=df['reviewText'].apply(lambda x: word_lemmatizer(x))
df['reviewText'].head()
```

```python
# stemming

from nltk.stem.porter import PorterStemmer
stemmer=PorterStemmer()
def word_stemmer(text):
    stem_text=[stemmer.stem(i) for i in text]
    return stem_text
```

```python
df['reviewText']=df['reviewText'].apply(lambda x: word_stemmer(x))
df['reviewText'].head(20)
```

```python
#since there is a chance that I include " summary " column in my model I am going to clean and process my summary coulmn as well, 
#as I did for the reviewText
```

```python
df['summary']=df['summary'].apply(lambda x: remove_punctuation(x))
df['summary']=df['summary'].apply(nltk.word_tokenize)
df['summary']=df['summary'].apply(lambda x: remove_stopwords(x))
df['summary']=df['summary'].apply(lambda x: word_lemmatizer(x))
df['summary']=df['summary'].apply(lambda x: word_stemmer(x))
```
